```python
# Calculate the mean values for the specified columns over the last 5 years
last_5_years = df[df['Year'].between(2015, 2020)]  # Filter data for the last 5 years
mean_co2_emissions = last_5_years['Value_co2_emissions_kt_by_country'].mean()
mean_electricity_fossil_fuels = last_5_years['Electricity from fossil fuels (TWh)'].mean()
mean_renewables_primary_energy = last_5_years['Renewables (% equivalent primary energy)'].mean()
mean_renewable_capacity_per_capita = last_5_years['Renewable-electricity-generating-capacity-per-capita'].mean()
mean_financial_flows = last_5_years['Financial flows to developing countries (US $)'].mean()
mean_electricity_renewables = last_5_years['Electricity from renewables (TWh)'].mean()
mean_renewable_share_total_energy = last_5_years['Renewable energy share in the total final energy consumption (%)'].mean()
mean_energy_intensity_primary_energy = last_5_years['Energy intensity level of primary energy (MJ/$2017 PPP GDP)'].mean()
mean_low_carbon_electricity = last_5_years['Low-carbon electricity (% electricity)'].mean()
mean_gdp_growth = last_5_years['gdp_growth'].mean()

# Fill missing values in the specified columns with the calculated means
df['Value_co2_emissions_kt_by_country'] = df['Value_co2_emissions_kt_by_country'].fillna(mean_co2_emissions)
df['Electricity from fossil fuels (TWh)'] = df['Electricity from fossil fuels (TWh)'].fillna(mean_electricity_fossil_fuels)
df['Renewables (% equivalent primary energy)'] = df['Renewables (% equivalent primary energy)'].fillna(mean_renewables_primary_energy)
df['Renewable-electricity-generating-capacity-per-capita'] = df['Renewable-electricity-generating-capacity-per-capita'].fillna(mean_renewable_capacity_per_capita)
df['Financial flows to developing countries (US $)'] = df['Financial flows to developing countries (US $)'].fillna(mean_financial_flows)
df['Electricity from renewables (TWh)'] = df['Electricity from renewables (TWh)'].fillna(mean_electricity_renewables)
df['Renewable energy share in the total final energy consumption (%)'] = df['Renewable energy share in the total final energy consumption (%)'].fillna(mean_renewable_share_total_energy)
df['Energy intensity level of primary energy (MJ/$2017 PPP GDP)'] = df['Energy intensity level of primary energy (MJ/$2017 PPP GDP)'].fillna(mean_energy_intensity_primary_energy)
df['Low-carbon electricity (% electricity)'] = df['Low-carbon electricity (% electricity)'].fillna(mean_low_carbon_electricity)
df['gdp_growth'] = df['gdp_growth'].fillna(mean_gdp_growth)
df['Access to clean fuels for cooking'] = df['Access to clean fuels for cooking'].ffill()  # Forward fill for Access to clean fuels for cooking

# Print the updated DataFrame with filled missing values
print("Added the Missing value with mean value taking for 5 year data")
print(df.head())
```

```python
from sklearn.linear_model import LinearRegression

# Initialize the linear regression model
model = LinearRegression()

# Fit the model using the training data
model.fit(X_train, y_train)
```

```python
# Make predictions using the model on the testing data
y_pred = model.predict(X_test)

# Evaluate the model's performance
from sklearn.metrics import mean_squared_error, r2_score

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate R-squared (coefficient of determination)
r_squared = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R-squared (Coefficient of Determination):", r_squared)
```
