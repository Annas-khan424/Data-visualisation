```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


# Selecting relevant features for clustering
X = df[['Value_co2_emissions_kt_by_country', 'Electricity from fossil fuels (TWh)', 'Renewables (% equivalent primary energy)']]

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

# Plotting the elbow method
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow Method to Determine Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors')

# Annotating the elbow point
plt.annotate('Elbow Point', xy=(3, sse[2]), xytext=(4, 1500),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Adding text with information
plt.text(6, 5000, 'Optimal K = 3', fontsize=12, bbox=dict(facecolor='yellow', alpha=0.5))

plt.show()


```

```python

# Define the optimal number of clusters
n_clusters = 3

# Initialize and fit the KMeans model
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Adding cluster labels to your DataFrame
df['Cluster'] = clusters

# Visualizing the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Value_co2_emissions_kt_by_country', y='Electricity from fossil fuels (TWh)', data=df, hue='Cluster', palette='viridis', s=100)
plt.title('Clusters Based on co2_emissions and Electricity from fossil fuels')
plt.show()
```

```python
alpha = 0.9  # Weight coefficient for CO2 Emissions
beta = 0.9   # Weight coefficient for Renewable Energy Adoption
gamma = 0.9  # Weight coefficient for Energy intensity level of primary energy (MJ/$2017 PPP GDP)

low_threshold = 0.3    # Threshold for low environmental impact category
high_threshold = 0.5   # Threshold for high environmental impact category

# Calculate Environmental Impact Score (Example)
df['Environmental Impact Score'] = (
    alpha * df['CO2 Emissions'] +
    beta * df['Renewable energy share in the total final energy consumption (%)'] +
    gamma * df['Energy intensity level of primary energy (MJ/$2017 PPP GDP)']
)

# Categorize countries based on Environmental Impact Score
df['Environmental Impact Category'] = pd.cut(
    df['Environmental Impact Score'],
    bins=[-np.inf, low_threshold, high_threshold, np.inf],
    labels=['Low Environmental Impact', 'Moderate Environmental Impact', 'High Environmental Impact']
)

# Plotting
impact_category_counts = df['Environmental Impact Category'].value_counts()
plt.figure(figsize=(8, 5))
impact_category_counts.plot(kind='bar')
plt.title('Countries Categorized by Environmental Impact')
plt.xlabel('Environmental Impact Category')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example data: scores for 3 different indicators across 10 observations
data = np.random.rand(10, 3)

# Initial weights
weights = np.array([0.3, 0.5, 0.2])

# Function to calculate composite score
def composite_score(data, weights):
    return np.dot(data, weights)

# Original composite scores
original_scores = composite_score(data, weights)

# Sensitivity analysis: vary one weight and adjust others proportionally
weight_ranges = np.linspace(0.1, 0.9, 9)
sensitivity_results = []

for w in weight_ranges:
    new_weights = np.array([w, 1.0-w/2, 1.0-w/2])
    new_weights /= new_weights.sum()  # Normalize weights
    new_scores = composite_score(data, new_weights)
    sensitivity_results.append(new_scores)

# Plotting results
for i, scores in enumerate(sensitivity_results):
    plt.plot(scores, label=f'Weights {np.round(weight_ranges[i], 2)}')

plt.title('Sensitivity Analysis of Composite Score to Weights')
plt.xlabel('Observation')
plt.ylabel('Composite Score')
plt.legend()
plt.show()
```
